```
import bs4 as bs  
import urllib.request  
```

# Getting text data

```
shinking_ratio = input("Shinking ratio (range 0-1, ex: 0.1): ")
shinking_ratio = float(shinking_ratio)
article_text = ""
keywords_showing = input("Showing key words: (0: False, 1:True) ") == "1"
summarizer_choice = input("Choosing summarizer: (0: Gensim, 1: Summa) ")

# Getting text data
if input("Getting text data from (0: web, 1: paste): ") == "0": # Web scraping
  url_topull = input('Enter the URL to pull - ')

  scraped_data = urllib.request.urlopen(url_topull)  
  article = scraped_data.read()
  parsed_article = bs.BeautifulSoup(article,'lxml')
  paragraphs = parsed_article.find_all('p')
  for paragraph in paragraphs:  
      article_text += paragraph.text
else:                                                          # Text pasting
  article_text = input("Paste your text:")
print("Data pull done")

# Summarization
summary = ""
if summarizer_choice == 0: # gensim summarization
  from gensim.summarization import summarize
  from gensim.summarization import keywords

  print("-----SUMMARY-----")
  summary = summarize(article_text, ratio=shinking_ratio)
  print (summary)
  print("-----------------")
  print()

  if keywords_showing:
    print("-----KEYWORDS-----")
    print (keywords(article_text, ratio=shinking_ratio))
    print("------------------")
    
else:                     # summa summarization
  from summa import summarizer
  from summa import keywords

  print("-----SUMMARY-----")
  summary = summarizer.summarize(article_text, ratio=shinking_ratio)
  print (summary)
  print("-----------------")
  print()

  if keywords_showing:
    print("-----KEYWORDS-----")
    print (keywords.keywords(article_text, ratio=shinking_ratio))
    print("------------------")
```

# Download summary

```
from google.colab import files

with open('summary.txt', 'w') as f:
  f.write(summary)
```
